# Estructuras de dataframes en detalle

Aprenderemos en detalle las diferentes variables que podemos encontrar en un mismo dataframe. Estas pueden ser categoricas, de tiempo, tipo texto, numericos (float e int).

**Una ventaja de trabajar con variables categoricas es que reducimos el tamaño de uso de la memoria RAM y el tamaño del archivo en si.**

## Variables categoricas

```
df_meteorites.nunique()
```

Vemos dos variables, **nametype** y **fall** que solo tiene dos tipos de datos. Esto lo podemos identificar como variable categorica.

```
df_meteorites[['fall', 'nametype']]
```

```
df_meteorites[['fall', 'nametype']].astype('category')
```

Para asegurarnos que el cambio se lleve a cabo dentro del dataframe hacemos lo siguiente:

```
df_meteorites[['fall', 'nametype']] = df_meteorites[['fall', 'nametype']].astype('category')
```

Ahora si vemos los tipos, podemos observar una nueva tipo de variable, **category**.

```
df_meteorites.dtypes
```

Si queremos saber los valores de la categoria lo podemos hacer llamando a la columna y con el metodo **unique()**.

```
df_meteorites['fall'].unique()
```

```
df_meteorites['nametype'].unique()
```

Si queremos saber la cantidad de valores categoricos que tenemos, o sea un conteo del total para cada categoria usamos el método **.value_counts()**

```
df_meteorites['fall'].value_counts()
```

```
df_meteorites['nametype'].value_counts()
```

Otra forma interesante de trabajar con variables categoricas es utilizar variables dummies.

```
pd.get_dummies(df_meteorites['fall'])
```

```
pd.get_dummies(df_meteorites['nametype'])
```

Convertir nuestras variables de formato texto a dummies es de gran importancia porque pasa de ser formato texto a formato numerico.

Podemos crear nuevas variables y utilizar el concepto anterior

```
df_meteorites['ones'] = 1
df_meteorites
```

```
df_meteorites[['fell', 'found']] = pd.get_dummies(df_meteorites['fall'])
df_meteorites
```

```
df_meteorites[['Relict', 'Valid']] = pd.get_dummies(df_meteorites['nametype'])
df_meteorites
```

## Variables de tiempo

```
df_meteorites['year']
```

Tenemos una variable tipo texto la cual nos interesaria transformarla en tipo tiempo, esto lo hacemos con el metodo **.to_datetime()**

```
pd.to_datetime(
    df_meteorites['year'],
    errors = 'coerce',
    format = '%m/%d/%Y %H:%M:%S %p'
)
```

Ahora guardamos lo anterior dentro de nuestro dataframe, sobreescribiendo la columna correspondiente.

```
df_meteorites['year'] = pd.to_datetime(
                                        df_meteorites['year'],
                                        errors = 'coerce',
                                        format = '%m/%d/%Y %H:%M:%S %p'
                                      )
```

Ahora podemos ver todas nuestras variables nuevamente reconvertidas.

```
df_meteorites.convert_dtypes().dtypes
```

Y ahora viendo como quedan las variables del dataframe puedo cambiar cambiar el nombre de alguna de las variables. 

```
df_meteorites.rename(columns={'mass (g)': 'mass'}, inplace=True)
```

Con el atributo inplace puedo mantener los cambios hechos anterirmente
